```python
f = h5py.File("/Users/yilinxia/Desktop/DXL/yt/chimera_000661800_grid_1_01.h5", "r")
# dataset can be donwloaded here http://use.yt/upload/6b106447
```

# Load JSON Schema 

```python
file = open('chimera_schema.json')
schema = json.load(file)
```

# Auto Function

```python
def auto_digest(hdf5, schema):
    dims = f[schema["geometry"]["dimensions"]][
        ()
    ]  # is this a gap between the manual and auto?
    data = {
        n: v[()].T
        for n, v in f[schema["fields"]["root"]].items()
        if v.shape == tuple(dims[::-1])
    }
    xgrid, ygrid, zgrid = (
        f[ax.replace(" ", "")][()]
        for ax in schema["geometry"]["hexahedral_mesh"]["edge"][1:-1].split(",")
    )
    coord, conn = yt.hexahedral_connectivity(xgrid, ygrid, zgrid)
    bbox = np.array(
        [
            [xgrid.min(), xgrid.max()],
            [ygrid.min(), ygrid.max()],
            [zgrid.min(), zgrid.max()],
        ]
    )
    ds = yt.load_hexahedral_mesh(
        data,
        conn,
        coord,
        bbox=bbox,
        geometry=(
            "spherical",
            tuple(
                schema["geometry"]["spherical"]["order"][1:-1]
                .replace(" ", "")
                .split(",")
            ),
        ),
    )
    return(ds)
```

# Auto Disgetion

```python
ds=auto_digest(f, schema)
```

```python
s = yt.SlicePlot(ds, "phi", ('stream','e_int'))
s.show()
```
